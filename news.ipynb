```python
from TextGenerator import TextGenerator
```

```python
text_gen = TextGenerator(method='n_grams', analyzer='word', n_grams=4)
```

```python
from nltk.tokenize import TweetTokenizer
tknzr = TweetTokenizer()

prepared_text = text_gen.prepare_for_genetation(data_folder_path='data/Lenta/', tokenizer=tknzr.tokenize, eos=['\n'])
```

```python
print('Number of words in corpus: ', len(prepared_text))
```

```python
text_gen.fit(prepared_text)
```

```python
print(text_gen.generate(generate_len=3000).replace('<sos>', '\n'))
```
